```python
## 1.IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##2.IMPORT DATASET
df=pd.read_csv("netflix_titles.csv")
pd.set_option('display.max_columns', None)
df.head(6)
```

```python
##3.DATA OVERVIEW
print(df.info)
```

```python
so data has about 8807 rows and 12 columns including-

1.show_id -Unique ID for each title

2.type- Movie or TV Show

3.title - Name of the content

4.director - Director of the movie/show

.cast - Main actors involved

6.country - Country of production

7.date_added -Date added to Netflix

8.release_year -Year the title was released

9.rating - Content maturity rating (e.g. TV-MA, PG-13)

10.duration - Length of the movie or number of seasons

11.listed_in - Genre or category (e.g. Drama, Comedy)

12.description - Short summary of the title

 so now we know the basics of the data we move onto the cleaning part to get it ready for our analysis.
```

```python
##DATA CLEANING
print(df.isna())
```

```python
So now lets see which columns have the most missing values to proceed for cleaning.
```

```python
print(df.isna().sum())
```

```python
 the columns of director, and countries have significant number of missing values,
So in order to perform our analysis we need to perform operations on rows in which-
1.ratings and date_added are missing (by removing them as they are too small in comparison with size of the dataset).
2.countries are missing(by replacing them with a "unknown" as the quantity is too large to ignore).
```

```python
df.dropna(subset=["rating","date_added"],inplace=True)
df["country"]=df["country"].fillna("unknown")
print(df.isna().sum())

```

```python
so now the data is cleaned for visualization and analysis.
```

```python
##EXPLORATORY ANALYSIS
so now lets proceed to what can we analyze and explore from our data-
1.-Which countryâ€™s productions are most represented on Netflix?
2.-comparison between number of shows vs number of movies.
3.-trends in number of uploads of movies and shows in netflix.
```

```python
##VISUALIZATIONS
Lets analyze the above listed queries one by one by visualizing and analyzing trends and insights.
-1.Which countryâ€™s productions are most represented on Netflix?
```

```python
#so to answer the question lets plot  a bar chart  of top 5 countries
top_countries=df["country"].value_counts().head(5)
plt.figure(figsize=(8,5))
plt.bar(top_countries.index,top_countries.values,color='tomato')
plt.title("Top 5 Countries with the Most Content on Netflix",fontsize=16,fontweight="bold")
plt.xlabel("countries")
plt.ylabel("number of titles")
plt.show
```

```python
#now lets look at the actual numbers
print(top_countries)
print(top_countries.sum())
```

```python
So, from the above bar graph and data table  as we can see the content from the united states is more than other three's countries content combine,also
if we include the unknown,the content from the united states is still more showing differences in content consumption of countries and global
content consumption in general.
```

```python
2.Now, letâ€™s compare the number of Movies and TV Shows on Netflix.
A bar chart is suitable here since it effectively represents category-wise comparisons.

```

```python
content_type=df["type"].value_counts()
plt.figure(figsize=(6,5))
plt.bar(content_type.index,content_type.values,color="lightgreen")
plt.title("Comparison between number of shows and number of movies",fontsize=14,fontweight="bold")
plt.xlabel("content type")
plt.ylabel("Number of Titles")
plt.show()

```

```python
print(content_type)
```

```python
The bar chart clearly shows that the number of movies available on Netflix is more than twice the number of TV shows.
This imbalance highlights Netflixâ€™s stronger focus on movies as the dominant content type in its catalog.
Factors influencing this trend may include shorter production timelines, lower costs, and higher audience reach compared to long-running TV shows.
It also reflects user behavior patterns â€” such as reduced attention spans and preference for quick, standalone entertainment â€” 
which make movies a more efficient form of content for both viewers and the platform.
```

```python
3.Now,lets explore the trends in number of uploads of content like movies and  tv shows yearwise.
we will here use a line chart which can effectively show any increase or decrease in values.
    
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
yearly_uploads=df["year_added"].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.plot(yearly_uploads.index,yearly_uploads.values,color="steelblue")
plt.title("Yearly Growth of Movies and TV Shows on Netflix",fontsize=14,fontweight="bold")
plt.xlabel("year")
plt.ylabel("Number of uploads")
plt.show()
```

```python
print((df['year_added'] == 2008).sum())
print((df['year_added'] == 2015).sum())
print((df['year_added'] == 2019).sum())



```

```python
The line chart clearly shows a sharp rise in Netflix uploads starting around 2015, which grew exponentially until 2019, followed by a slight decline.
During this four-year period, the number of titles added to Netflix increased from about 73 to nearly 2,000.
This surge reflects factors such as the expansion of internet access in developing countries, the rapid global growth of Netflix, 
and its strategic push to diversify and localize content before and during the pandemic years.
```
